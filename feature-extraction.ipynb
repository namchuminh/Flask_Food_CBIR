# Tải mô hình ResNet50 (Feature extraction)

```python
import numpy as np
import pandas as pd
import os
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics.pairwise import cosine_similarity
from tensorflow.keras.applications.resnet50 import preprocess_input
from keras.preprocessing import image
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Kích thước ảnh đầu vào 150x150
img_width, img_height = 150, 150

# Khởi tạo mô hình ResNet50 và bỏ đi lớp fully connected (top layer)
model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))
```

# Xây dựng hàm trích xuất đặc trưng

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.utils import load_img, img_to_array
# Hàm trích xuất đặc trưng
def extract_features(image_path):
    # Load ảnh và resize
    img = load_img(image_path, target_size=(img_width, img_height))
    # Chuyển ảnh thành mảng numpy
    img_array = img_to_array(img)
    # Mở rộng chiều để phù hợp với đầu vào của mô hình (batch_size, height, width, channels)
    img_array = np.expand_dims(img_array, axis=0)
    # Tiền xử lý ảnh
    img_array = preprocess_input(img_array)
    # Trích xuất đặc trưng
    features = model.predict(img_array, verbose=0)
    # Làm phẳng mảng đặc trưng
    return features.flatten()
```

# Huấn luyện & đánh giá

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Giả sử bạn có nhãn y_train tương ứng với X_train
# Chia X_train và y_train thành tập huấn luyện và kiểm tra tạm thời
X_train_part, X_val_part, y_train_part, y_val_part = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

# Khởi tạo và huấn luyện mô hình KNN
num_similar_items = 5
knn_model = NearestNeighbors(n_neighbors=num_similar_items, metric='cosine')
knn_model.fit(X_train_part)

# Đánh giá mô hình trên tập kiểm tra tạm thời
def precision_at_k(y_true, y_pred, k):
    """Tính Precision tại k"""
    y_true_k = y_true[:k]
    y_pred_k = y_pred[:k]
    return np.sum(np.isin(y_pred_k, y_true_k)) / k

def mean_average_precision(y_true_all, y_pred_all, k=10):
    """Tính Mean Average Precision cho toàn bộ tập kiểm tra"""
    average_precisions = []
    for y_true, y_pred in zip(y_true_all, y_pred_all):
        precision_k = precision_at_k(y_true, y_pred, k)
        average_precisions.append(precision_k)
    return np.mean(average_precisions)

# Dự đoán và tính Precision@k, MAP trên tập kiểm tra tạm thời
y_true_all = []
y_pred_all = []

for x, y_true in zip(X_val_part, y_val_part):
    distances, indices = knn_model.kneighbors([x])
    y_pred = [y_train_part[i] for i in indices[0]]  # Nhãn dự đoán từ tập huấn luyện

    y_true_all.append([y_true])  # Ground truth là nhãn thực sự
    y_pred_all.append(y_pred)   # Nhãn dự đoán

# Đánh giá Precision@10 và MAP
k = 5
precision = precision_at_k(y_true_all[0], y_pred_all[0], k)
map_score = mean_average_precision(y_true_all, y_pred_all, k)

print(f"Precision@{k}: {precision:.4f}")
print(f"Mean Average Precision (MAP): {map_score:.4f}")
```

```python
def recall_at_k(y_true, y_pred, k):
    """Tính Recall tại k"""
    y_true_set = set(y_true[:k])
    y_pred_set = set(y_pred[:k])
    return len(y_true_set & y_pred_set) / len(y_true_set) if y_true_set else 0

def average_precision(y_true, y_pred, k=10):
    """Tính Average Precision (AP) cho mỗi ảnh trong tập kiểm tra"""
    precision_scores = []
    for i in range(1, k + 1):
        if y_pred[i - 1] in y_true:
            precision_i = precision_at_k(y_true, y_pred, i)
            precision_scores.append(precision_i)
    return np.mean(precision_scores) if precision_scores else 0.0

# Tính Recall@10 và Average Precision (AP) cho tập kiểm tra tạm thời
k = 5
recall_scores = []
ap_scores = []

for y_true, y_pred in zip(y_true_all, y_pred_all):
    recall = recall_at_k(y_true, y_pred, k)
    ap = average_precision(y_true, y_pred, k)
    
    recall_scores.append(recall)
    ap_scores.append(ap)

# Tính Recall trung bình và Average Precision trung bình
mean_recall = np.mean(recall_scores)
mean_ap = np.mean(ap_scores)

print(f"Recall@{k}: {mean_recall:.4f}")
print(f"Average Precision (AP): {mean_ap:.4f}")
```
